```python
import numpy as np
```

```python
def predicted_model_means(iter_1, iter_2, stat_1, stat_2):
    if iter_1 == 0:
        iter_1 += 1
        return 0
    if iter_2 == 0:
        iter_2 += 1
        return 1
    model_1 = stat_1 / iter_1
    model_2 = stat_2 / iter_2

    if model_1 < model_2:
        return 1
    else:
        return 0
```

```python
def execute_model(j, iter_1, iter_2, stat_1, stat_2):
    if j == 0:
        stat_1 += round(np.random.uniform(0, 1))
        iter_1 += 1
    else:
        stat_2 += round(np.random.uniform(0, 1))
        iter_2 += 1

    return stat_1, stat_2, iter_1, iter_2
```

```python
def greedy(epochs=10, mode='eps'):
    eps = 0.2
    iter = 0
    iter_1, iter_2, stat_1, stat_2 = 0, 0, 0, 0
    while iter < epochs:
        iter += 1
        p = np.random.uniform(0, 1)
        if mode == 'eps':
            if p < eps:
                print('choose random')
                j = round(np.random.uniform(0, 1))
            else:
                print('choose best')
                j = predicted_model_means(iter_1, iter_2, stat_1, stat_2)
        else:
            j = predicted_model_means(iter_1, iter_2, stat_1, stat_2)
            
        stat_1, stat_2, iter_1, iter_2 = execute_model(j, iter_1, iter_2, stat_1, stat_2)
        print(f'iter_1: {iter_1}, stat_1: {stat_1}, iter_2: {iter_2}, stat_2: {stat_2}')
        
        if iter_1 == 0:
            model_1 = 0
        else:
            model_1 = stat_1 / iter_1

        if iter_2 == 0:
            model_2 = 0
        else:
            model_2 = stat_2 / iter_2
        
        print(f'iter: {iter} model 1: {model_1}, model 2: {model_2}')
```

```python
greedy(epochs=100, mode='eps')
```
